# CSV Manipulation

This is a sample program to demonstrate some common CSV file manipulations for data cleaning.
1. First we read a CSV file and extract some row and columns to create 2 x-y plots; one for the number of agricultural related bachelor degrees conferred from 1970-2022, and one for the number of computer information related bachelor degrees conferred from 1970-2022.
2. Then we create a function with parameters x, y and iterate through every column and row to produce a table of all the number of bachelor degrees conferred by year.

```python
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("jobs.csv")

# Remove all commas from original dataframe.
df = df.replace(',','', regex=True)

# See the first 19 rows.
df.head(19)
```

# Making a table for agriculture degrees conferred by year from 1970-2022

```python
# Convert the main pandas dataframe to a numpy object.
table = df.to_numpy()

# Segment the years list and agriculture list to their respective variables.
years = table[0][1:]
agriculture = table[2][1:]

# Cast the agriculture numpy array as int.
agriculture = agriculture.astype(int)

print(years)
print(agriculture)

# Create a plot only for years as x and agriculture & natural resource degrees conferred from 1970-2021.
plt.figure(figsize=(20,10), dpi=200)
plt.plot(years, agriculture, "ob")
plt.xlabel('Years')
plt.ylabel('Agriculture & Natural Resources')
plt.title('Bachelor\'s degrees conferred by postsecondary institutions: Selected academic years, 1970-71 through 2021-22')
```

# Making a table for computer information degrees conferred by year from 1970-2022

```python
# Convert the main pandas dataframe to a numpy object.
table = df.to_numpy()

# Segment the years list and agriculture list to their respective variables.
years = table[0][1:]
computation = table[9][1:]

print(table[2][1:]) # first subject
print(table[33][1:]) # last subject
print()

print(table[2][1]) # first element from first subject
print(table[2][18]) # last element from first subject
print()

# Cast the agriculture numpy array as int.
computation = computation.astype(int)

print(years)
print(computation)

# Create a plot only for years as x and computer information science degrees conferred from 1970-2021.
plt.figure(figsize=(20,10), dpi=200)
plt.plot(years, computation, "ob")
plt.xlabel('Years')
plt.ylabel('"Computer and information sciences and support services"')
plt.title('Bachelor\'s degrees conferred by postsecondary institutions: Selected academic years, 1970-71 through 2021-22')
```

# Now we're creating an automated script that prints out every chart of bachelor degrees conferred by subject at once.

```python
def makePlot(years, degrees_conferred, subject):
    plt.figure(figsize=(10,5), dpi=150)
    plt.plot(years, degrees_conferred, "ob")
    plt.xlabel("years")
    plt.ylabel(subject)
    plt.title('Bachelor\'s degrees conferred by postsecondary institutions: Selected academic years, 1970-71 through 2021-22')
```
